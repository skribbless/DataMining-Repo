```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
dg['Month'] = pd.Categorical(dg['Month'], categories=month_order, ordered=True)
```

```python
monthly_cases = dg.groupby([dg.Month, dg.Year])["Dengue_Cases"].sum()
```

```python
monthly_cases.unstack().plot(kind='line')
plt.gcf().set_size_inches(12,8)
plt.grid(True)
plt.xticks(range(0,12), month_order)
plt.show()
```

<h1>Temporal Trends of Dengue Incidents</h1>

```python

data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
        'Cases': [5, 12, 8, 7, 10]}
df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Cases'], marker='o', color='b')
plt.title('Dengue Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

<h1>Insight 2</h1>
<p1>Correlation Between Rainfall and Dengue Incidents</p1>

```python
data = {'Rainfall': [10, 20, 30, 40, 50], 'DengueCases': [5, 15, 20, 25, 30]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Rainfall', y='DengueCases', data=df)
plt.title('Correlation Between Rainfall and Dengue Cases')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Number of Dengue Cases')
plt.show()

# Calculate correlation
correlation = df['Rainfall'].corr(df['DengueCases'])
print(f'Correlation between Rainfall and Dengue Cases: {correlation}')
```

<h1>Insight 3</h1>
<p1>Effectiveness of Public Health Interventions</p1>

```python
data = {'Period': ['Before', 'After', 'Before', 'After', 'Before'],
        'Cases': [50, 30, 60, 40, 55]}

df = pd.DataFrame(data)
sns.boxplot(x='Period', y='Cases', data=df)
plt.title('Dengue Cases Before and After Public Health Intervention')
plt.show()
```

<h1>Insight 4</h1>
<p1>Dengue deaths per year</p1>

```python
df = pd.read_csv('datasets//dengue.csv')

df.head()
df_grouped = df.groupby('Year')['Deaths'].sum().reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Deaths', data=df_grouped, palette='viridis')

plt.title('Dengue Deaths per Year')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()

print(df_grouped.describe())
```
